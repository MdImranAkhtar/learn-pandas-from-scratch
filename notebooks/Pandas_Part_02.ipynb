```python
# Reading the 'deliveries.csv' file into a DataFrame named 'delivery'
delivery = pd.read_csv('/content/deliveries.csv')

# Reading the 'matches.csv' file into a DataFrame named 'match'
match = pd.read_csv('/content/matches.csv')
```

```python
# Displaying the first 5 rows of the 'delivery' DataFrame

delivery.head()
```

```python
# Grouping the 'delivery' DataFrame by the 'batsman' column and storing the result in the variable 'run'

run = delivery.groupby('batsman')
```

```python
# Retrieving all delivery records for the batsman 'V Kohli'

run.get_group('V Kohli')
```

```python
# Calculating the total runs scored by each batsman,
# sorting them in descending order,
# and displaying the top 10 highest run-scorers

run['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Creating a mask to filter rows where the batsman scored exactly 6 runs (i.e., hit a six)
mask = delivery['batsman_runs'] == 6

# Applying the mask to create a new DataFrame containing only sixes
new_delivery = delivery[mask]
```

```python
# Getting the total number of sixes hit (i.e., number of rows in new_delivery)

new_delivery.shape[0]
```

```python
# Counting the number of sixes hit by each batsman,
# sorting them in descending order,
# and displaying the top 5 batsmen with the most sixes

new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
# Creating a new DataFrame 'vk' that contains all delivery records where the batsman is 'V Kohli'

vk = delivery[delivery['batsman'] == 'V Kohli']
```

```python
# Calculating the total runs scored by V Kohli against each bowling team,
# sorting the teams in descending order of runs conceded,
# and displaying the top 5 teams he scored the most against

vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# Defining a function to calculate the top 3 teams against which a given batsman has scored the most runs
def run_scored(batsman_name):

    # Filtering the deliveries DataFrame for the specified batsman
    vk = delivery[delivery['batsman'] == batsman_name]

    # Grouping by bowling team, summing runs, sorting in descending order, and returning top 3
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# Calling the run_scored function for 'V Kohli' to get the top 3 teams
# he has scored the most runs against

run_scored('V Kohli')
```

```python
# Calling the run_scored function for 'MS Dhoni' to get the top 3 teams
# he has scored the most runs against

run_scored('MS Dhoni')
```

#**The  isin() function**

```python
#isin()
#find the most destructive death over batsman in the history of ipl
#Strike Rate=(Runs/Balls)*100
#Min batsman 200 balls in over 16-20
```

```python
# Creating a new DataFrame 'death_over' containing deliveries from the death overs (overs 16 to 20)

death_over = delivery[delivery['over'] > 15]
```

```python
# Counting the number of balls faced by each batsman in death overs
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()

# Creating a mask to filter batsmen who have faced more than 200 balls in death overs
x = all_batsman > 200

# Displaying only those batsmen who have faced more than 200 balls in death overs
all_batsman[x]
```

```python
# Converting the index of filtered batsmen (who faced more than 200 balls in death overs)
# into a list and storing it in 'batsman_list'

batsman_list = all_batsman[x].index.tolist()
```

```python
# Displaying the list of batsmen who have faced more than 200 balls in death overs

batsman_list
```

```python
#Runs scored by all these 43 batsman
#Balls played by these 43 batsman

final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
# Calculating the total runs scored by each of the 43 selected batsmen

runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
# Counting the total number of balls faced by each of the 43 selected batsmen

balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
# Calculating the strike rate for each batsman using the formula: (total runs / total balls) * 100

sr = (runs / balls) * 100
```

```python
# Sorting the strike rates of batsmen in descending order to see who has the highest

sr.sort_values(ascending=False)
```

# **The Merge function**

```python
# Merging the 'delivery' and 'match' DataFrames based on match IDs
# 'match_id' column from 'delivery' is matched with 'id' column from 'match'

delivery.merge(match, left_on='match_id', right_on='id')
```

```python
# Storing the merged DataFrame of 'delivery' and 'match' into a new DataFrame named 'new'
# This allows access to both ball-by-ball and match-level information in one place

new = delivery.merge(match, left_on='match_id', right_on='id')
```

```python
# Printing the shape (number of rows and columns) of the 'delivery' DataFrame
print(delivery.shape)

# Printing the shape (number of rows and columns) of the 'match' DataFrame
print(match.shape)
```

```python
# Grouping the merged DataFrame 'new' by 'season' and 'batsman' to calculate total runs per season per batsman
# Sorting the result to get the top scorer per season
# Resetting index to turn grouped columns into regular columns
# Dropping duplicates to keep only the top scorer for each season
# Sorting the final result by season in ascending order

new.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')
```

#**Pivot Table**

```python
# Creating a mask to filter rows where the batsman scored 6 runs (i.e., hit a six)
mask = delivery['batsman_runs'] == 6

# Applying the mask to create a new DataFrame 'six' containing only the sixes
six = delivery[mask]

# Displaying the first 5 rows of the 'six' DataFrame
six.head()
```

```python
# Displaying the shape of the 'six' DataFrame to see the total number of sixes and columns

six.shape
```

```python
# Creating a pivot table to count the number of sixes hit by each batting team in each over
# Rows represent overs, columns represent batting teams, and values are the count of sixes

pt = six.pivot_table(index='over', columns='batting_team', values='batsman_runs', aggfunc='count')
```

```python
# Importing the seaborn library for data visualization, especially for statistical plots

import seaborn as sns
```

```python
# Creating a heatmap to visualize the number of sixes hit by each team in each over
# Darker colors represent higher counts of sixes

sns.heatmap(pt)
```
